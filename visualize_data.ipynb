```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pickle

%matplotlib inline
```

```python
ep_length = 400
sim_steps = 50000
```

```python
save_data = pickle.load(open("data/data.p", "rb"))
input_x = save_data["x"] # (49999, 7)
input_y = save_data["y"] # (49999, 2)
x = input_y[:400, 0]
z = input_y[:400, 1]
```

```python
def create_colors(length, segments):
    c_list = []
    for i in range(segments):
        color = np.zeros([length//segments, 4])
        color[:, 3] = 0.5
        color[:, 0:3] = np.random.uniform(0, 1, size=[1, 3])
        c_list.append(color)
    c_list = np.reshape(c_list, [length, 4])
    return c_list
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(input_y[:400, 0], input_y[:400, 1], s=1, c=create_colors(sim_steps, sim_steps//ep_length))
plt.ylim(-4.5, 4.5)
plt.xlim(-4.5, 4.5)
```

Point around (0.5, -1) is the start position, and right top pos is the end position.

```python
plt.figure(figsize=(10, 10))
plt.scatter(input_y[:, 0], input_y[:, 1], s=1, c=create_colors(sim_steps, sim_steps//ep_length))
plt.ylim(-4.5, 4.5)
plt.xlim(-4.5, 4.5)
```
